3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

    - 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

        - count: Counter 인스턴스가 지니는 초기 정수 값
        - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
                    >>> c = Counter(10)

    - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

                    >>> print(c)

                    [Count (step: 2)] 10 
    
    - 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

                    >>> c.incr() 
                    >>> d.incr() 
                    >>> print(c) 
                    [Count (step: 1)] 11 
                    >>> print(d) 
                    [Count (step: 2)] 12 
                    
    - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

                    >>> c() 
                    >>> d() 
                    >>> print(c) 
                    [Count (step: 1)] 12 
                    >>> print(d) 
                    [Count (step: 2)] 14 
    - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

                    >>> c = c + 5 
                    >>> d = d - 5 
                    >>> print(c) 
                    [Count (step: 1)] 17 
                    >>> print(d) 
                    [Count (step: 2)] 9 
                    
    - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

                    >>> print(c) > 10 
                    True 
                    >>> print(d) > 10 
                    False 
                    >>> print(c) < 10 
                    False 
                    >>> print(d) < 10 
                    True 
                    >>> print(c) == 17 
                    True 
                    >>> print(d) != 9 
                    False 

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return str('[Count (step: ' + str(self.step) + ')] ' + str(self.count))
    
    def incr(self):
        self.count = self.count + self.step
    
    def __call__(self):
        self.incr()
        
    def __add__(self, A):
        self.count = self.count + A
        return self
    
    def __sub__(self, A):
        self.count = self.count - A
        return self
    
    def __gt__(self, A): # greater than
        return self.count > A
    
    def __lt__(self, A): # less than
        return self.count < A
    
    def __eq__(self, A): # equal
        return self.count == A
    
    def __ne__(self, A): 
        return self.count != A
```

```python
if __name__ == "__main__":
    c = Counter(10)
    d = Counter(10, 2)
    print("요구사항 1&2")
    print(c)
    print(d)
    print()
    
    print("요구사항 3")
    c.incr()
    d.incr()
    print(c)
    print(d)
    print()
    
    print("요구사항 4")
    c()
    d()
    print(c)
    print(d)
    print()
    
    print("요구사항 5")
    c = c + 5
    d = d - 5
    print(c)
    print(d)
    print()
    
    print("요구사항 6")
    print(c > 10)
    print(d > 10)
    print(c < 10)
    print(d < 10)
    print(c == 17)
    print(d != 9)
```

#### [문제 4]
4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

    - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

- 1. \__init__\()
    - 생성자가 호출될 시 실행되는 함수이다. MySet 클래스는 리스트의 서브 클래스로 append를 사용할 수 있다. append는 listd에 append의 인자로 들어온 값을 list에 추가한다. 따라서 'for e in l:'에서 l의 원소를 하나씩 e로 넣은 뒤 이를 list의 서브 클래스인 MySet 클래스의 객체 인스턴스에 하나씩 저장한다. 그런 뒤 정적 메소드 eliminate_duplicate를 클래스 이름을 사용해 호출해 객체 인스턴스에 중복된 원소를 제거해 준다.


- 2. \__str__\()
    - 객체 인스턴스를 print() 함수를 이용할 때 print()의 괄호 안이 str() 과정을 거쳐 \__str__함수가 호출된다. 객체 인스턴스 내의 원소를 특정 형태로 출력하기 위해 result에 'MySet:'을 저장하고 그 뒤에 for문을 써 객체 인스턴스의 요소 하나씩 ','와 함께 result에 연속적으로 이어지도록 저장했다. 마지막 원소가 출력될 때 같이 출력된 ','를 제거하기 위해 [0:len(result)-2]까지 슬라이싱하여 result에 저장하였다. 그런 뒤 이를 반환하였다.


- 3. eliminate_duplicate()
    - 클래스 이름 공간에 존재하는 정적 메소드이다. 다른 곳에서 호출할 때 클래스 이름을 이용해 호출한다. 이 함수는 인자로 리스트를 받아 빈 리스트에 s에 for문, if문의 과정을 거쳐 l에서 리스트의 요소 값을 하나씩 e에 넣어 e와 같은 값이 s에 존재하지 않으면 s에 추가하고, 중복된 값은 빼고 s에 저장한다. l을 슬라이싱을 통해 비우고 s의 값들을 l에 집어넣어 최종적으로 중복된 원소가 없는 리스트를 만들어준다.

#### [문제 5]
5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

     - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
    - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

                    >>> u = s | t 
                    >>> print(u) 
                    MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
                
    - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

                    >>> u = s & t 
                    >>> print(u) 
                    MySet: {2, 3} 
                
    - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

                    >>> s = MySet([1, 2, 3]) 
                    >>> t = MySet([3, 4, 5]) 
                    >>> u = s - t 
                    >>> print(u) 
                    MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    # s에 self의 리스트 요소를 넣은 후 반복문을 통해 A의 리스트 요소중 s에 없는 값은 추가
    # sort()를 사용해 리스트 s를 정렬
    # 반환할 때 MySet 클래스의 객체 인스턴스로 반환해야하므로 
    # a = MySet(s)를 통해 객체 인스턴스 a가 리스트 s를 인자로 가지도록 하고 a를 반환
    def __or__(self, A):
        s = []
        for e in self:
            s.append(e)
        for e in A:
            if e not in s:
                s.append(e)
        s.sort()
        a = MySet(s)
        return a
    
    # s에 self의 리스트 요소들 중 A 리스트에 존재하면 추가
    # sort()를 사용해 리스트 s를 정렬
    # MySet 생성자의 인자로 넣고 만들어진 객체 인스턴스 반환
    def __and__(self, A):
        s = []
        for e in self:
            for i in A:
                if e == i:
                    s.append(e)
        s.sort()
        a = MySet(s)
        return a
    
    # s에 self의 리스트 요소들 중 A 리스트에 존재하지 않으면 추가
    # sort()를 사용해 리스트 s를 정렬
    # MySet 생성자의 인자로 넣고 만들어진 객체 인스턴스 반환
    def __sub__(self, A):
        s = []
        for e in self:
            if e not in A:
                s.append(e)
        s.sort()
        a = MySet(s)
        return a
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    print()
    
    print("요구사항 1")
    u = s | t
    print(u)
    print()
    
    print("요구사항 2")
    u = s & t
    print(u)
    print()
    
    print("요구사항 3")
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```

```python
import requests
import string

class SearchEngine:
    
    def __init__(self,*args):
        self.urllist = list(args)
        self.dicts = []
    
    def addUrl(self,url):
        self.urllist.append(url)
    
    def removeUrl(self,url):
        if url in self.urllist:
            del self.urllist[self.urllist.index(url)]
    
    def listUrls(self):
        for url in self.urllist:
            print(url)
    
    def getWordsFrequency(self):
        self.dicts = []
        srclist = []
        
        if self.urllist == []:
            print({})
            return {}
        
        for i in range(len(self.urllist)):
            srclist.append(requests.get(self.urllist[i]).text)
            
        for i in range(len(self.urllist)):
            Punctuation = self.remove(self.search(srclist[i])) 
            dic = self.countWords(Punctuation)
            self.dicts.append(dic)
            
        return self.dicts
    
    def getMaxFrequencyWords(self):
        self.getWordsFrequency()
        maxWords = []
        
        if self.urllist == []:
            return None
        print()
        
        for i in range(len(self.urllist)):
            maxWordsUrl = []
            result = []
            list1 = list(self.dicts[i].keys())
            list2 = list(self.dicts[i].values())
            Tuple = tuple(zip(list1,list2))
            list2.sort(reverse=True) 
            
            # 비교했을 때 같은 경우 - key값을 넣어 빈도수가 가장 높은 단어 순으로 result에 추가
            for j in range(len(list2)):
                for k in range(len(Tuple)):
                    if list2[j] == Tuple[k][1]: 
                        result.append(Tuple[k][0])
            print(self.urllist[i])
            
            for j in range(len(list2)-1):
                maxWordsUrl.append(result[j])
                print(result[j])
                if list2[j] != list2[j+1]:
                    break;
            maxWords.append(maxWordsUrl)
        return maxWords
    
    def searchUrlByWord(self,word):
        self.getWordsFrequency()
        similarity = [0 for i in range(len(self.urllist))]
        slist = word.split()
        slist = self.search(slist)
        
        # 이미 있는 단어인 경우 - similarity에 더함
        # similarity가 가장 높게 나온 사이트 먼저 
        for i in range(len(self.urllist)): 
            for j in slist:
                for k in self.dicts[i].keys():
                    if j == k:
                        similarity[i] += self.dicts[i][j]
        list = []
        
        tempDict = {}
        for i in range(len(self.urllist)):
            tempDict[self.urllist[i]]=0

        for i in range(len(self.urllist)):
            list.append(similarity[i])
        temp = [i for i in similarity]
        temp.sort(reverse=True)
       
        # value 값이 1이면 출력 X
        # value 값이 1이 아니면 다음 url 확인
        for i in range(len(temp)):
            for j in range(len(list)):
                if temp[i] == list[j]:
                    if tempDict[self.urllist[j]] == 0:
                        print("유사도:", list[j], "=>", self.urllist[j])
                        tempDict[self.urllist[j]]=1
                        if (i<len(list)-1) and (temp[i] != temp[i+1]) :
                            return
                        break
                    elif tempDict[self.urllist[j]] == 1:
                        continue
    

    def search(self,src):
        i = 0
        text = ''
        while i < len(src): 
            if src[i] == '<': 
                if src[i:i+7] == '<script':
                    i += 7
                    while src[i:i+9] != '</script>':
                        i += 1
                    i += 9
                elif src[i:i+6] == '<style': 
                    i += 6
                    while src[i:i+8] != '</style>':
                        i += 1
                    i += 8
                else :   
                    i += 1
                    while src[i] != '>':
                        i += 1
                    i += 1
            else:
                text += src[i]
                i += 1
        return text.split()
    
    # 사전으로 변환
    def Filter(self,list):
        with open("E_stopwords.txt",'r') as f:
            E_words = f.read()
        with open("Kor_stopwords.txt",'r') as f:
            Kor_words = f.read()
        with open("sign.txt",'r') as f:
            signWords=f.read()

        E_stopwords = E_words.split() 
        Kor_stopwords = Kor_words.split('\n') 
        sign_words = signWords.split() 


        # 영어 불용어 필터
        for i in range(len(list)):
            for j in range(len(E_stopwords)):
                if list[i].lower() == E_stopwords[j]:
                    list[i] = ''

        # 한글 불용어 필터 
        for i in range(len(list)):
            for j in range(len(Kor_stopwords)):
                if len(list[i]) >= len(Kor_stopwords[j]):
                    if list[i][-len(Kor_stopwords[j]):] == Kor_stopwords[j]:
                        list[i] = list[i][:-len(Kor_stopwords[j])]


        # 특수문자 필터
        for i in range(len(list)):
            for j in range(len(sign_words)):
                list[i]=list[i].replace(sign_words[j],'')

        return list
    
    # 특수문자 및 원소 값이 비어있으면 제거
    def remove(self,list):
        for i in range(len(list)):
            for j in range(len(string.punctuation)):
                list[i] = list[i].replace(string.punctuation[j],'')

        list = self.Filter(list) # 불용어 제거

        i = 0
        while i != len(list):
            if list[i] == '':
                del list[i]
                continue
            i += 1
        return list
    
    # 단어들의 출현빈도
    def countWords(self,list):
        d = {}
        for i in range(len(list)):
            d[list[i]] = 0
        for i in range(len(list)):
            d[list[i]] = d[list[i]] + 1
        return d
    
if __name__ == '__main__':
    w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
    w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com') 
    w3 = SearchEngine() 
    
    w1.addUrl('https://github.com')
    w3.addUrl('http://stackoverflow.com')
    
    w1.removeUrl('http://www.cnn.com')
    w2.removeUrl('http://stackoverflow.com')
    
    w2.listUrls()

    w1.getWordsFrequency() # 반환
    w1.getMaxFrequencyWords() # 결과 출력
    
    print()
    w1.searchUrlByWord("news")
```
